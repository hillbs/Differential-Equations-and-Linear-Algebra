###Notes

* The large numbers constructing a life table are never going to be outweighed by the small numbers of multiple comorbidities. Hence, it is probably reasonable to allow the life table to be seen as $h_0$ due to the neglibible influence of the small comorbidity population on the overall population, even though there is in effect double counting.
* Rather than ignoring the $h_0$ term, we are making it explicit as the lifetable hazard rate. By the Law of Large Numbers, this is a useful assumption.
* There also is the benefit that the $h_0$ term is out to around 60+ years largely linear so that the underlying assumption of constant hazard rate is mostly preserved. 
* We also can add to this a way of incorporating baselines from other datasets. Need to think this through again as to why. If the baselines are similar enough, probably no need. If not, then to join coefficients, there needs to be the same baseline. This can be done by a regression (possibly using PyStan) of the epidemiological dataset (e.g., SEER, Framingham) and incorporating the baseline from that set into the main new hazard rate as below. 
\begin{equation} h(t) = h_{ baseline}(t) e^{(\beta_0 + \vec{\beta_1} \vec{X_{CI}})} \ h_{ study} e^{(\vec{\beta_s} \vec{X_s})}\      (1.)\end{equation}
* The above equation has the flaw that we only know $h(t)$ and $h_{baseline} $. To get to the CI indices, we would need to convert these using a regression equation, as below.

* How to calculate 10-year survival...
    * Calculate Charlson Score or Index (i)
    * Add Comorbidity score to age score
    * Total denoted as 'i' below
    * Calculate Charlson Probablity (10 year mortality)
    * Calculate Y = e^(i * 0.9)
    * Calculate Z = 0.983^Y where Z is the 10 year survival
    * from http://www.fpnotebook.com/prevent/Exam/ChrlsnCmrbdtyIndx.htm



A test of Eq.1 is whether the predicted mortality rate $h(t)$ is derived from $\vec X_{CI}$ as the CI alone. In fact it turns our to be more proper to have the exponent in the form $\beta_0 + \beta_1 \vec X_{CI}$.

```python
import statsmodels.api as sm
X = range(7)
X = sm.add_constant(X)
y = log((mortality)/df.ix[:,4].mean())
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
CI_mort = {i:(exp(-1.5422 + .5911*i))*df.ix[:,4].mean() for i in range(7)}
CI_mort
```

So this test doesn't really validate Eq.1 at all. Really, all we are doing with the CI_mort variable is approximating the parameters using OLS; we could equally well just use the original data as the $h(t)$ value. Or, possibly a Bayesian estimate of parameters if we had a ton more data.
