```python
import matplotlib.pyplot as plt
```

Data from:
    http://www.hsmr.be/wp-content/uploads/publicaties/Charlson%20Comorbidity%20index%20JCE%202004.pdf

```python
raw = "69.4 70.2 76.2 75.9 15.7 15.3 11.7 11.8 7.7 7.6 6.4 6.3 3.3 3.0 2.2 2.3 1.3 1.2 0.8 0.9 1.8 2.1 2.1 2.1 0.8 0.8 0.6 0.7"
raw = array(raw.split(),dtype=float32).reshape(7,4)
y = raw.mean(axis=1)
raw,y
```

```python
plt.bar([0,1,2,3,4,5,6],y, 
        color='g')
plt.grid(True)
plt.title('Distribution of Charlson Index Scores',
          size = 16)
plt.xlabel('Charlson Score: Disease burden')
plt.ylabel('Percentage(%) in each range')
y, raw.sum(axis=0)
```

```python
raw2 = "0.3 0.4 0.4 2.5 3.4 3.2 5.0 6.6 5.6 9.0 11.6 10.7 12.8 14.9 13.6 14.9 16.3 14.7 21.1 24.8 23.6"
raw2 = array(raw2.split(), dtype=float32).reshape(7,3)
mortality = raw2.mean(axis=1)
plt.bar([0,1,2,3,4,5,6],mortality, 
        color='r',
        alpha=.4)
plt.grid(True)
plt.title('Distribution of Charlson Index Scores',
          size = 16)
plt.xlabel('Charlson Score: Disease burden')
plt.ylabel('Percentage(%) in-hospital mortality')
raw2,raw2.sum(axis=0)
```

Not sure why the sum of the percentage mortality only adds to 65.6, 78.0 and 71.8. So the plot below is multiplying the probability of being in a particular Charlson score range x the probability of then dying if you are in that range. This would work, except the probabilities in both need to sum to 1.0. If I normalised the mortality percentages this would increase the absolute range of mortality but not change the relationships between each. Maybe the plot above already is the conditional mortality, $P(Death| Charlson Score = x)$

```python
mortality_rate = (raw.mean(axis=1).T/100*raw2.mean(axis=1)/100.)*1e5
plt.bar(xrange(7),mortality_rate,
        alpha=.5)
plt.title('Mortality rate')
plt.xlabel('Charlson score')
plt.ylabel('Mortality x $10^5$')
plt.grid(True)
```
