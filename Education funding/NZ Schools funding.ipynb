```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
funding = pd.read_excel('/Users/LlewelynRW/Downloads/Operational-Funding-to-Schools-2001-2014.xlsx',
                        "A5",header=5,parse_cols="B:P",skip_footer=3)
#funding = funding.ix[:,1:]*1000
funding = funding.T
funding.columns = list(funding.ix[0])
funding = funding.drop(funding.index[0])
funding
```

```python
Fig1 = funding.plot(figsize=(12,8),fontsize=12, title="Funding levels by Decile")
Fig1.set_xlabel('Year',fontsize=14)
Fig1.set_ylabel('x $1000',fontsize=14)
Fig1.grid()
```
