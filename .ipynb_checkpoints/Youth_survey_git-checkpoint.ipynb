### Ethnicity

```python
df.ix[df.NZEuro >= 1,'NZEuro'] = 'NZEuro'
df.ix[df.Maori >= 1,'Maori'] = 'Maori'
df.ix[df.Samoan >= 1,'Samoan'] = 'Samoan'
df.ix[df.Cook_Is >= 1,'Cook_Is'] = 'Cook_Is'
df.ix[df.Tongan  >= 1,'Tongan'] = 'Tongan'
df.ix[df.Niuean  >= 1,'Niuean'] = 'Niuean'
df.ix[df.Chinese  >= 1,'Chinese'] = 'Chinese'
df.ix[df.Indian  >= 1,'Indian'] = 'Indian'

df['Ethnicity'] = df['NZEuro']\
.add(df.Maori, fill_value=' ')\
.add(df.Samoan,fill_value=' ')\
.add(df.Cook_Is,fill_value=' ')\
.add(df.Tongan,fill_value=' ')\
.add(df.Chinese,fill_value=' ')\
.add(df.Indian,fill_value=' ')\

df = df.drop(['NZEuro','Maori','Samoan','Cook_Is','Tongan','Niuean','Chinese','Indian'],axis=1)

df.replace(['NZEuroMaoriSamoan    ','NZEuro  Cook_Is   ','NZEuroMaori     ','NZEuro Samoan    ', ' MaoriSamoan    ',
       ' Maori Cook_Is   ', 'NZEuroMaori Cook_Is   ','NZEuro     Indian',  ' SamoanCook_Is   ','NZEuroMaoriSamoanCook_Is   '],\
           ['Multiple','Multiple','Multiple','Multiple','Multiple','Multiple','Multiple','Multiple','Multiple','Multiple'],inplace=True)
df.head()
```

```python
Ethnic_dist = df.Ethnicity.value_counts()
pd.Series(df.Ethnicity.ravel()).unique()
```

```python
N = Ethnic_dist.count()
ind = np.arange(N)  # the x locations for the groups
width = .8       # the width of the bars
f = plt.figure(facecolor=".9")
ax = f.add_subplot(111, axisbg="#F4EAEA")
rects1 = ax.bar(ind, Ethnic_dist.values[0:N], width, color='purple')

# add some text for labels, title and axes ticks
ax.set_ylabel('$Frequency$',fontsize=14)
ax.set_title('$Ethnic \ Distribution$',fontsize=20)
ax.set_xticks(ind+width/3)
xtickNames = ax.set_xticklabels( pd.Series(df.Ethnicity.ravel()).unique() )
plt.setp(xtickNames, rotation=75, fontsize=14)
def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*height, '%d'%int(height),
                ha='center', va='bottom')

autolabel(rects1)
plt.grid(True)
plt.show()
```

```python
#Create function to plot bar chart
def Make_bar(A,B,C,Header, Y_label):
    '''A is <object>.count()
       B is <object.values[0:A]
       c is <object>.keys()
       Header is plot title
       Y_Label is y-axis title'''
    N = A#10#schools.count()
    ind = np.arange(N)  # the x locations for the groups
    width = .8       # the width of the bars
    f = plt.figure(facecolor=".9")
    ax = f.add_subplot(111, axisbg="#F4EAEA")
    rects1 = ax.bar(ind, B, width, color='purple',alpha=.4)

    # add some text for labels, title and axes ticks
    ax.set_ylabel(Y_label,fontsize=14)
    ax.set_title(Header,fontsize=20)
    ax.set_xticks(ind+width/3)
    xtickNames = ax.set_xticklabels(C)
    plt.setp(xtickNames, rotation=75, fontsize=14)
    def autolabel(rects):
        # attach some text labels
        for rect in rects:
            height = rect.get_height()
            ax.text(rect.get_x()+rect.get_width()/2., 1.05*height, '%d'%int(height),
                    ha='center', va='bottom')

    autolabel(rects1)
    plt.grid(True)
    plt.show()
    return
```
