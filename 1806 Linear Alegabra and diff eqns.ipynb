The above plots and formulae are illustrating the impact of the combination of a growth and oscillation combined. The point of this is to make the damping ratio simply a function of two units, $\omega_n^2 = \frac{C}{A}$, which is the natural frequency and $p$. The damping ratio, $\zeta = \frac{B}{\sqrt{4AC}}$. Oscillation occurs, now, as a function of $\zeta$ as a ratio around 1. The idea is that there is impact (damping, oscillation) when the natural frequency $\omega_n$ is changed by $p^2$, where the result is $\omega_d^2 = \omega_n^2 - p^2 = \omega_n^2(1 - \zeta^2)$. 

Strang in pp. 103-104 outlines an important method for solving forced oscillations with an exponential response, as above. For the equation $Ay'' + By' + Cy = e^{st}$, the solution $y = y_n + y_p.$ We have $ y_n = C_1e^{s_1t} + C_1e^{s_2t}$. And $y_p = Ye^{st}.$ One technique is to substitute this term into the original equation, cancel the $e^{st}$ terms and solve for $Y$. Note of course that you need to apply the differentiation for each term. This process can be shortened by using a transfer function. For these type of equations, s is determined from the $f(t) = e^{st}$, and $Y$ becomes $ \frac {1}{As^2 + Bs + C}$. The final solution then becomes $ y = Ye^{st} = \frac {1}{As^2 + Bs + C}e^{st}$. In sympy, the dsolve function solves for most of these $Ly = f(t)$ equations.

### 2.7 Laplace Transforms

p.135 of Strang describes $y(t)$ as a growth factor $g(t-T)$ multiplying the inputs $f(T)$. When each input is summed over a continuous range, this sum then is the integral of the product. That is, $y_p(t) = \int_0^{t} g(t-T)f(T)dT$. Note the comparison with the Laplace Transform, whose function is to change terms around integration into algebraic problems. 

$f(t) \rightarrow F(s) =  \int_0^{t} f(t)e^{-st}dt$

Sympy and python provide the following functions to perform the transforms:

```python
from sympy.integrals import *
a,b,s = symbols('a b s', positive =True)
laplace_transform(exp(a*t),t,s)
```

An important extension is to calculate the transform of the derivative, $\frac {dy}{dt}$.

```python
D = diff(exp(a*t),t)
L = laplace_transform(D,t,s)
L
```

Note how the coefficient $a$ now multiplies the original transform of $e^{at} = Y(s)$ to $aY(s)$. 

```python
f = Eq(diff(g(t),t)-a*g(t),exp(b*t))
laplace_transform(f,t,s).evalf()
#The result should be...
f, 1/((s-a)*(s-b))
```

```python
F = 1/((s-a)*(s-b))
inverse_laplace_transform(F,s,t)
#see pp.141-142 for an explanation. 
```

Now, lets illustrate how the Laplace transform works, including using partial fractions and sympy commands on Example 1, p.140. 

$y'' - 4y' + 3y = e^{at}$

```python
#Using the laplace transform function.
E = Eq(diff(g(t),t,2)-4*diff(g(t),t)+3*g(t),exp(a*t))
laplace_transform(E,t,s)
```

Sympy just doesn't handle abstract functions that well and we end with an unevaluated expression. 

A really useful general reference for how to solve partial fractions is found at: http://mattpap.github.io/scipy-2011-tutorial/html/partfrac.html

```python
#So the alternative is to manually calculate the Laplacian.
F = integrate(exp(-s*t)*g(t),(t,0,oo))
F = F.subs(g(t),exp(a*t)).doit()
simplify(F)
```

```python
F = integrate(exp(-s*t)*diff(g(t),t),(t,0,oo))
F = F.subs(g(t),exp(a*t)).doit()
F = simplify(F)
F
```
