## Simple (one spatial dimension) Heat or Diffusion Equation

```python
Heat = Eq(diff(u(x,t),t), k*diff(u(x,t),x,2))
Heat
```

```python
Solution = Eq(u(x,t),(1/(sqrt(4*pi*k*t)))*Integral(exp((x-y)**2)/(4*k*t)*V3(y),(y,-oo,oo)))
Solution
```

```python
%matplotlib inline
pylab.rcParams['figure.figsize'] = 12, 12
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()

ax = fig.gca(projection='3d')
X, T = np.mgrid[0:6:.1, .2:50:1]
Z = f(X,T)

surf = ax.plot_surface(X,T , Z, cmap='jet', cstride=2, rstride=2)
ax.set_xlabel("Distance (x)")
ax.set_ylabel("Duration (t)")
ax.set_zlabel("u(t,x)")
#ax.set_zlim(Z.min, Z.max())

plt.show()
```
