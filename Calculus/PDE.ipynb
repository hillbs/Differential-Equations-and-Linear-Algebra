##### Video 6:

```python
PDE6 = Eq(diff(u(x,t),t)+diff(u(x,t),x)+7*u(x,t),0)
PDE6
```

```python
classify_pde(PDE6)
```

```python
pdsolve(PDE6)
```

```python
l=f(x-t)*exp(-7*x)
checkpdesol(PDE6,l)
```

So the general solution on the video and the Python one equate.

```python
#The general solution 
l
```

```python
#From the Cauchy parameters, u(x,0) = e^x
Soln = Eq(l.subs(t,0),exp(x))
Soln
```

```python
#Obtain f(x)
Solnb = Eq(Soln.lhs/exp(x*-7),Soln.rhs/exp(-7*x))
Solnb
```

```python
#But x actually is x-t, so substitute back in.
Solnc = Solnb.subs(x,x-t)
Solnc
```

```python
#And we know that u(x,t) = f(-t+x)*e^-7x, so substitute for f(-t+x) with result above, simplify, and obtain the final result!
Final_6 = l.subs(f(-t+x),Solnc.rhs)
Eq(u(x,t),simplify(Final_6))
```

##### Video 16: Solving a quasi-linear PDE using the method of charactieristics.

```python
#The problem.
Q = Eq(x*(x**2+u(x,y,t))*diff(u(x,y,t),x)-y*(x**2+u(x,y,t))*diff(u(x,y,t),y),(x**2-y**2)*u(x,y,t))
Q
```

```python
#what sort of equation is it?
try:
    classify_pde(Q)
except NotImplementedError:
    print('Currently Sympy cannot solve this type of PDE')
```

```python
#so lets use the method of characteristics
CE = Eq(diff(x(t),t)/(x*(x**2+u(x,y))),Eq(diff(y(t),t)/(-y*(x**2+u(x,y))),diff(u(x,y,t),t)/((x**2-y**2)*u(x,y,t))))
CE
```

```python
#Try using Python to simplify -- still no luck
simplify(CE)
```

```python
#Manually rewrite
CE = Eq(diff(x(t),t)/x*(x**2+u(x,y)),Eq(diff(y(t),t)/y*-(x**2+u(x,y)),diff(u(x,y,t),t)/u(x,y,t)*(x**2-y**2)))
CE
```

```python
#Add first two terms, note that we multiply by -1/-1 to try and equate denominator.
CE = Eq(diff(x(t),t)/x*(x**2+u(x,y))+diff(y(t),t)/y*-(x**2+u(x,y)),-1*diff(u(x,y,t),t)/(u(x,y,t)*(-1*(x**2-y**2))))
CE
```

```python
simplify(CE)
```

## Wave Equations: Using d'Alembert's equation and Duhanel's principle to solve inhomogenous Equations

This is based on material from Videos 17-24.

```python
#The Wave Equation
init_session()
c,s = symbols('c s')
V1, V2 = symbols('phi psi', cls=Function)
W = Eq(diff(u(x,t),t,2),c**2*diff(u(x,t),x,2))
W
```

```python
#The general solution is...
G = Eq(u(x,t), f(x+c*t)+g(x-c*t))
G
```

```python
#d'Alembert and Duhanel give the following general solution.
G= Eq(u(x,t), 1/2*(V1(x+c*t)+V1(x-c*t))+1/(2*c)*Integral(V2(s),(s,x-c*t,x+c*t))+1/(2*c)*Integral(Integral(h(y,s), (y,(x-c*(t-s)),(x+c*(t-s)))),(s,0,t)))
G
```

```python
#Define the Wave equation
W = Eq(diff(u(x,t),t,2)-c**2*diff(u(x,t),x,2),x).subs(c,1)
W
#this form cannot be solved by any current Sympy solver.
```

```python
#Cauchy boundaries
phi=x**3
psi=cos(x)
phi, psi
```

```python
#Rewrite the function phi as x**3.
G= Eq(u(x,t), 1/2*((x+c*t)**3+(x-c*t)**3)+1/(2*c)*Integral(V2(s),(s,x-c*t,x+c*t))+1/(2*c)*Integral(Integral(h(y,s), (y,(x-c*(t-s)),(x+c*(t-s)))),(s,0,t)))
G
# Substitute c =1, for h, and for psi.
G = G.subs(c,1).subs(h(y,s),x).subs(V2(s),psi)
G
```

```python
G = G.doit()
G
```

```python
simplify(G)
```
