# RWC analysis

The above sections read the data from the RWC site, save the dfs and clean the data 

```python
Team_Stats = pd.read_pickle('/Users/LlewelynRW/Dropbox/Calculus materials/18.086/Sport_Analytics/Team_stats.pkl')
PoolA = pd.read_pickle('/Users/LlewelynRW/Dropbox/Calculus materials/18.086/Sport_Analytics/PoolA.pkl')
PoolB = pd.read_pickle('/Users/LlewelynRW/Dropbox/Calculus materials/18.086/Sport_Analytics/PoolB.pkl')
PoolC = pd.read_pickle('/Users/LlewelynRW/Dropbox/Calculus materials/18.086/Sport_Analytics/PoolC.pkl')
PoolD = pd.read_pickle('/Users/LlewelynRW/Dropbox/Calculus materials/18.086/Sport_Analytics/PoolD.pkl')
Team_Stats
```

```python
PoolB.ix[1,0] = 'SOUTH AFRICA'
PoolB.ix[1,1] = 0
PoolC.ix[1,0] = 'NEW ZEALAND'
PoolC.ix[1,1] = 0
PoolC
```

```python
RWC15 = pd.concat([PoolA,PoolB,PoolC,PoolD]) 
RWC15['Rank'] = 1.+ (RWC15.W-RWC15.L)/2.
RWC15.TEAM,Team_Stats.TEAM
```

```python
#Create a normally distributed strength for each team
Team_Stats['PTS_Per_Game'] = Team_Stats.PTS/Team_Stats.PLD
Team_Stats['Strength'] = (Team_Stats.PTS_Per_Game - Team_Stats.PTS_Per_Game.mean())/Team_Stats.PTS_Per_Game.std()

#Make the joining column compatible.
Team_Stats.TEAM = map(str.upper,Team_Stats.TEAM)
RWC_Combined = pd.merge(Team_Stats,RWC15,on='TEAM').sort('Strength',ascending=False)
#Set the index to the team name.
RWC_Combined = RWC_Combined.set_index(['TEAM'])
RWC_Combined.Strength
```

```python
Standard_background()
RWC_Combined.Strength.plot(kind='bar',title='RWC 2015',figsize=(12,8),color='purple',table=False
                           ,fontsize=12)
plt.grid(True)
```
