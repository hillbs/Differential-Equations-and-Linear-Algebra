```python
#Create function to plot bar chart
def Make_bar(A,B,C,Header, Y_label,color='purple',alpha=.7):
    '''A is <object>.count()
       B is <object.values[0:A]
       c is <object>.keys()
       Header is plot title
       Y_Label is y-axis title'''
    N = A#10#schools.count()
    ind = np.arange(N)  # the x locations for the groups
    width = .8       # the width of the bars
    f = plt.figure()#facecolor=".9")
    ax = f.add_subplot(111)#, axisbg="#F4EAEA")
    rects1 = ax.bar(ind, B, width, color=color,alpha=alpha)

    # add some text for labels, title and axes ticks
    ax.set_ylabel(Y_label,fontsize=16)
    #ax.set_xlabel(ind,fontsize=16)
    ax.set_title(Header,fontsize=20)
    ax.set_xticks(ind+width/3)
    ax.set_fontsize=16
    xtickNames = ax.set_xticklabels(C)
    plt.setp(xtickNames, rotation=45, fontsize=12)
    ymax = np.max(B)*1.1
    plt.ylim(ymax=ymax) 
    def autolabel(rects):
        # attach some text labels
        for rect in rects:
            height = rect.get_height()
            ax.text(rect.get_x()+rect.get_width()/2., 1.05*height, '%d'%int(height),
                    ha='center', va='bottom',fontsize=16)

    autolabel(rects1)
    plt.grid(True)
    plt.show()
    return
```

```python
print 'The personal concerns raised were as a percentage of all participants:'
Concerns8 = df.loc[:,'Lonely':'Other.3'].count()/len(df.index)*100
print Concerns8
print
print 'The second rating of general concerns:'
Concerns9 = df.loc[:,'Lonely.1':'Other.4'].count()/len(df.index)*100
print Concerns9
Concerns_diff = Concerns8.sub(Concerns9.values)
```

```python
Concerns_list = ['Lonely','Relationships','Grades','D & A','Bullying','Lack of $',
                 'Violence','Finding a job','Identity','Culture','Friends','Family','Other']
Make_bar(Concerns8.count(),Concerns8.values[0:Concerns8.count()],Concerns_list,
         'What matters to me: Overall','Percentage',color='magenta')
```

```python
Make_bar(Concerns9.count(),Concerns9.values[0:Concerns9.count()],Concerns_list,
         'What matters to others: Overall','Percentage',color='cyan')
```

```python
#Concerns_diff.plot(kind='bar')
Make_bar(Concerns_diff.count(),Concerns_diff.values[0:Concerns_diff.count()],Concerns_list,
         'Difference between what matters to me versus others','Percentage change',color='lime')
```

The above plot shows the differential between concerns that are personal to the youth compared to how that person rated general concerns. So, for example, $5$% rated drugs and alcohol as a personal issue. However, this was rated as a general issue by $54$%. This $-48$% difference indicates that more than half of youth perceive drugs and alcohol as a major issues where only $5$% rated it as personally impacting on them. Similarly, bullying and violence were also rated as more relevant than personally experienced. This probably indicates general concerns, perhaps some expectation and stereotyping.  

Conversely, school grades were rated as much more personally than generally concerning. So one might expect that this indicates an issue with personal self efficacy wherein youth feel grades, or the lack thereof, to be more applicable to themselves than to their peers. In any event, grades present a general concern for youth. The last differential of note was the issue of loneliness, with $14$% of youth rating themselves as feeling lonely, but with a perception that this was a slightly lesser issue for others. What is interesting is that the differential wasn't larger, suggesting that youth are aware of others being lonely. So the question arising is that where this impression is relatively accurate, that still the issue of loneliness persists. Being lonely and being aware of other lonely people doesn't appear to assist in changing anything.

The above ratings were interesting for the fact that the ratings suggest youth want to involved LESS in half of the areas rated. For example, $45$% indicated current involvement with physical activity. However, only $11$% indicated that they wanted to be involved in physical activity. The same pattern was evident for Sport, Music, School ($17$%) Shopping, Hangin with friends and Hangin with Whanau/Family. There was a slight desire to be more involved with environmental and community spirit type activities. These responses were quite intriguing. Do they mean that youth are searching for activities with greater meaning than simply "fun" activities? It would be interesting and probably significant for organisations seeking to meet youth needs to question whether this finding represents inaccurate fundamental assumptions being made on their part about youth needs. A followup might be a focus group with high school youth to ask them what they think is going on with this result.
