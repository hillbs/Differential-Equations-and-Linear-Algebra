```python
#Create function to plot bar chart
def Make_bar(A,B,C,Header, Y_label,color='purple',alpha=.7):
    '''A is <object>.count()
       B is <object.values[0:A]
       c is <object>.keys()
       Header is plot title
       Y_Label is y-axis title'''
    N = A#10#schools.count()
    ind = np.arange(N)  # the x locations for the groups
    width = .8       # the width of the bars
    f = plt.figure()#facecolor=".9")
    ax = f.add_subplot(111)#, axisbg="#F4EAEA")
    rects1 = ax.bar(ind, B, width, color=color,alpha=alpha)

    # add some text for labels, title and axes ticks
    ax.set_ylabel(Y_label,fontsize=16)
    #ax.set_xlabel(ind,fontsize=16)
    ax.set_title(Header,fontsize=20)
    ax.set_xticks(ind+width/3)
    ax.set_fontsize=16
    xtickNames = ax.set_xticklabels(C)
    plt.setp(xtickNames, rotation=45, fontsize=14)
    ymax = np.max(B)*1.1
    plt.ylim(ymax=ymax) 
    def autolabel(rects):
        # attach some text labels
        for rect in rects:
            height = rect.get_height() 
            if np.min(B) < 0:
                ax.text(rect.get_x()+rect.get_width()/2., 0, '%d'%int(height),
                        ha='center', va='bottom',fontsize=16)
            else:
                ax.text(rect.get_x()+rect.get_width()/2., 1.05*height, '%d'%int(height),
                        ha='center', va='bottom',fontsize=16)

    autolabel(rects1)
    plt.grid(True)
    plt.show()
    return
```

```python
#Concerns_diff.plot(kind='bar')
Make_bar(Concerns_diff.count(),Concerns_diff.values[0:Concerns_diff.count()],Concerns_list,
         'Difference between what matters to me versus others','Percentage change',color='lime')
```

The above plot shows the differential between concerns that are personal to the youth (what matters to them) compared to how that person rated what issues youth in Eastside face generally. So, for example, $5$% rated drugs and alcohol as a personal issue. However, this was rated as a general issue by $54$%. This $-48$% difference indicates that more than half of youth perceive drugs and alcohol as a major issue for Eastside Youth, even though only $5$% rated it as personally impacting on them. Similarly, bullying and violence were also rated as more relevant than personally experienced. This probably indicates general concerns, perhaps some expectation and stereotyping.  

Conversely, school grades were rated as much more personally than generally concerning. So one might expect that this indicates an issue with personal self efficacy wherein youth feel grades, or the lack thereof, to be more applicable to themselves than to their peers. In any event, grades present a general concern for youth. The last differential of note was the issue of loneliness, with $14$% of youth rating themselves as feeling lonely, but with a perception that this was a slightly lesser issue for others. What is interesting is that the differential wasn't larger, suggesting that youth are aware of others being lonely. So the question arising is that where this impression is relatively accurate, that still the issue of loneliness persists. Being lonely and being aware of other lonely people doesn't appear to assist in changing anything.

### Bullying and who it impacts upon

```python
df.Bullying.fillna(0,inplace=True)
table6 = pd.pivot_table(df, index=['School'], columns=['Bullying'], values=['ID'],
                        aggfunc=count_nonzero,margins=False)
#Calculate percentages
table6 = np.round(table6.apply(lambda x : x / x.sum()*100., axis=1),2)
table6
```

```python
# as a reference http://pbpython.com/pandas-pivot-table-explained.html
table6.plot(kind='barh',width=.7,alpha=.4,color=['w','r'],hatch='//',fontsize=14)
plt.legend(labels=['No report of bullying', 'Reports being bullied'])
plt.title('Bullying reporting by School',fontsize=20)
plt.xlabel('Percentage',fontsize=16)
plt.ylabel('School',fontsize=16)
plt.grid(True)
```

```python
#Bullying by Gender
df.Bullying.fillna(0,inplace=True)
table7 = pd.pivot_table(df, index=df['Gender_(M/F)'], columns=['Bullying'], values=['ID'],
                        aggfunc=count_nonzero,margins=False)
#Calculate percentages
table7 = np.round(table7.apply(lambda x : x / x.sum()*100., axis=1),2)
table7
```

```python
table7.plot(kind='barh',width=.7,alpha=.4,color=['w','r'],hatch='//',fontsize=14)
plt.legend(labels=['No report of bullying', 'Reports being bullied'])
plt.title('Bullying reporting by Gender',fontsize=20)
plt.xlabel('Percentage',fontsize=16)
plt.ylabel('Gender',fontsize=16)
plt.grid(True)
```

Whilst bullying is often gauged by the impression of adults, a strength of these data is that youth were asked to rate whether it was an issue for them. They also were asked to indicate if they were concerned at it being an issue for Eastside youth in general. 

Taking the issue of personal experience, as reported by youth, it is interesting that the distribution of bullying across schools varies. Also, it seems that, as youth progress to Intermediate, the perception of bullying increases, but, by high school, has dropped to its lowest percentage. At Intermediate, this is the only place where girls feel more bullied than boys (7% more).

It would be fascinating to ascertain what type of bullying and the perceived impact on youth as a follow up. 

There were no clear gender differences between boys or girls being bullied, although there was a tendency for girls to report this slightly more.
