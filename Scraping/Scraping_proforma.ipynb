```python
url = 'http://yellow.co.nz/rotorua/coffee?what=coffee&where=Rotorua'
        #raw_input("Enter the url: ")
r  = requests.get(url)
data = r.content
soup = BeautifulSoup(data)

for link in soup.find_all('a'):
    print(link.text,link.get('href'))
```

```python
#finding strtings that are in the html
import bs4
def print_if_email(s):
    if 'http' in s: print s

for div in soup.findAll('div'):
    # Iterate over the children of each matching div
    for c in div.descendants:
        # If it wasn't parsed as a tag, it may be a NavigableString
        if isinstance(c, bs4.element.NavigableString):
            # Some heuristic to identify email addresses if other non-tags exist
            print_if_email(c)
```
