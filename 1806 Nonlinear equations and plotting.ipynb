So the vector field plots rely on being able to provide a gradient vector for the function $f(x,y)$. The gradient vector is $< \frac {df}{dx}(x,y,z), \frac {df}{dy}(x,y,z), \frac {df}{dz}(x,y,z)>$. So, in using the quiver plot function, the values for a 2D plot, $U, V$ need to be specified in these terms. The exception of course is when the slope, only depends on one variable, $X$ or $Y$, in which instance the gradient already is noted. 

```python
%matplotlib inline
Y, X = np.mgrid[-3:3:15j, -5:5:15j]
U = X #the x or t value
V = Y-Y**2 # the slope
speed = np.sqrt(U*U + V*V)

plt.streamplot(X, Y, U, V, color=U, linewidth=2, cmap=plt.cm.autumn)
plt.contour(X,Y,Y-Y**2)
plt.colorbar()
plt.grid()

f, (ax1) = plt.subplots(ncols=1)
ax1.quiver(X, Y, U, V)
ax1.grid(b=True, which='major', color='b', linestyle='-', alpha=0.5)                                                
ax1.grid(b=True, which='minor', color='g', linestyle='-',alpha = .2) 
plt.minorticks_on()

plt.show()
```

The above is the streamplot and vector plot of the function $\frac{dy}{dt} = Y - y^2$.

The trick, or hard part, is to find values for $U$ and $V$, where one is the base value and the other the slope. 

```python
Y, X = np.mgrid[-2:2:15j, -2:2:15j]
U = Y-1 #the x or t value
V = 1+X # the slope
speed = np.sqrt(U*U + V*V)

plt.streamplot(X, Y, U, V, color=U, linewidth=2, cmap=plt.cm.autumn)
plt.contour(X,Y,1+X-Y)
plt.colorbar()
plt.grid()

f, (ax1) = plt.subplots(ncols=1)
ax1.quiver(X, Y, U, V)
ax1.grid(b=True, which='major', color='b', linestyle='-', alpha=0.5)                                                
ax1.grid(b=True, which='minor', color='g', linestyle='-',alpha = .2) 
plt.minorticks_on()
ax1.contour(X,Y,1+X-Y)
plt.show()
```

This example is $\frac{dy}{dt} = 1 + t - y$, where one sees the zero isocline, $f(t) = 1 + t - y = 0$, being solved by the line $y = 1 + t$.

```python
Y, X = np.mgrid[-pi:pi:15j, -pi:pi:15j]
U = 2*X #the slope wrt X
V = 2*Y # the slope wrt Y
speed = np.sqrt(U*U + V*V)

plt.streamplot(X, Y, U, V, color=U, linewidth=2, cmap=plt.cm.autumn)
plt.colorbar()
plt.grid()
Z = X**2+Y**2
f, (ax1) = plt.subplots(ncols=1)
ax1.contour(X, Y, Z)
ax1.quiver(X,Y,U,V)
ax1.grid(b=True, which='major', color='b', linestyle='-', alpha=0.5)                                                
ax1.grid(b=True, which='minor', color='g', linestyle='-',alpha = .2) 
plt.minorticks_on()

plt.show()
```

The streamplot for $y' = \frac {-x}{y}$ is above. The isoclines are circles for each constant, $c$, that solves the equation $\frac{-x}{y} = c$. The contour function provides the plot of isoclines.

### 3.3 Linearisation and stability

The logistic equation $y' = y - y^2 $, with $Y = 0,\ Y = 1$. Stability requires $\frac{df}{dy} < 0 \ at \ Y$.

```python
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
Z = Y-Y**2
U = X #the x or t value
V = Z#Z is the slope of Y already without calculating gradient. 
levels = np.arange(1,3, 1)
plt.streamplot(X, Y, U, V, color=U, linewidth=1, cmap=plt.cm.autumn)
plt.colorbar()
plt.grid()

f, (ax1) = plt.subplots(ncols=1)
ax1.contour(X, Y, Z,levels,linewidth=3)
ax1.streamplot(X, Y, U, V, color='g', linewidth=1, cmap=plt.cm.autumn)
ax1.grid(b=True, which='major', color='b', linestyle='-', alpha=0.5)                                                
ax1.grid(b=True, which='minor', color='g', linestyle='-',alpha = .2) 
plt.minorticks_on()

plt.show()
```

The nonlinear autonomous equations have $\frac {dy}{dt} = f(y,z) \ and \frac{dz}{dt} = g(y,z)$ starting from $y(0)$ and $z(0)$.

Where these two functions equal 0, there are same 6 possibilities:

Stable: Sink $(T<0,\ D>0)$ $(T^2 \geq 4D)$ or Spiral sink $(T<0,\ D>0)$ $(T^2 < 4D)$ . 

Neutral: Centre $(T=0,\ D>0)$ $(T^2 < 4D)$ .

Unstable: Source $(T>0,\ D>0)$ $(T^2 \geq 4D)$ , Spiral source $(T>0,\ D>0)$ $(T^2 < 4D)$ , Saddle point $(D<0)$ $(T^2 \geq 4D)$ .

 The  idea is to linearise $f(y,z)$ and $g(y,z)$. This is like the Taylor series whre the first term is set to 0, except in two dimensions. It brings out the Jacobian. Whe the value $Y$ is $Y(0)$, then it produces the equation with traces and determinents. 

```python
a,b,c,d,y,z = symbols('a b c d y z')
prey = y-y*z
pred = y*z - z
pred,prey
```

```python
#Predator prey relationship
Y, X = np.mgrid[0:3:100j, 0:3:100j]
Z = Y-Y**2
U = X - X*Y #slope of f; time is now out of the equation to preserve 2d.
V = Y*X - Y#slope of g
plt.streamplot(X, Y, U, V, color=U, linewidth=1, cmap=plt.cm.autumn)
plt.colorbar()
plt.grid()
plt.show()
```

```python
#Manually calculation the Jacobian and applying to the linearised equations.
f1=prey
f2=pred
f1y=diff(f1,y)
f1z=diff(f1,z)
f2y=diff(f2,y)
f2z=diff(f2,z)
J = Matrix([[f1y,f1z],[f2y,f2z]])
J, J.subs([(y,0), (z,0)])
```

```python
#Using the inbuilt Jacobian function. Evaluate at saddle point (0,0).
P = Matrix([prey,pred])
P.jacobian([y,z]).subs([(y,0), (z,0)])
```

```python
#Using the inbuilt Jacobian function. Evaluate at neutral point (1,1).
P = Matrix([prey,pred])
P.jacobian([y,z]).subs([(y,1), (z,1)])
```

```python
#You can find lambda by just factoring the characteristic poly, saving computing time (4x). 
lam = symbols('lambda')
P1 = P.jacobian([y,z]).subs([(y,1), (z,1)])
factor(P1.charpoly(lam))
```

```python
#Or use the eigenvects function, which when solving for lamda is 2/3 of time of a solve call.
P1.eigenvects()
```

```python
solve(factor(P1.charpoly(lam)),lam)
```

Now lets re-do this equation for a more realistic sert of inputs and where coefficients are not unitary. 

```python
a,b,c,d,y,z,p,q,r,s,w = symbols('a b c d y z p q r s w')
prey = y*(p-q*y-r*z)
pred = z*(-s+w*y)
pred,prey
```

First we need to solve for when $f(y,z) = g(y,z)$. This is done using the solve function, with the equation passed as a tuple, indicating equality and that it is a linear equation. 

```python
Solns = solve((y*(p-q*y-r*z),z*(-s+w*y)),y,z)
simplify(Solns)
```

```python
P = Matrix([prey,pred])
P0 = P.jacobian([y,z]).subs([(y,Solns[0][0]), (z,Solns[0][1])])
P0, P.jacobian([y,z]), factor(P0.charpoly(lam))
```

```python
P1 = P.jacobian([y,z]).subs([(y,Solns[1][0]), (z,Solns[1][1])])
P1, P.jacobian([y,z]), factor(P1.charpoly(lam))
```

```python
P3 = P.jacobian([y,z]).subs([(y,Solns[2][0]), (z,Solns[2][1])])
simplify(P3), P.jacobian([y,z]), factor(P3.charpoly(lam)).doit()
```

Notice that the original solve call creates solutions equating $f(y,z) = g(y,z)$ for x and y. Inspecting the equations also suggests solutions for when $p$ and $s$ are 0. This can be implemented as below:

```python
Solns2 = solve((y*(p-q*y-r*z),z*(-s+w*y)),s,p)
Solns2
```

```python
P4 = P.jacobian([y,z]).subs([(p,Solns2[p]), (s,Solns2[s])])
P4, P.jacobian([y,z]), factor(P4.charpoly(lam)).doit()
```

```python
#Now lets plot these equations!
Y, X = np.mgrid[0:3:100j, 0:3:100j]
p,q,r,s,w = 3.,1.,2.,3.,5.
U = X*(p-q*X-r*Y) #slope of f; time is now out of the equation to preserve 2d.
V = Y*(-s+w*X)#slope of g
plt.streamplot(X, Y, U, V, color=U, linewidth=1, cmap=plt.cm.autumn)
plt.colorbar()
plt.grid()
plt.show()
```
