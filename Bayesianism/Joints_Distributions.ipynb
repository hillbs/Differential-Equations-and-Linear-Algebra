# Joint distributions

Probably the more useful manner to understand the joint pdf model is plotting it. Seabon seems to offer some useful plotting functions to do this. http://stanford.edu/~mwaskom/software/seaborn/tutorial/distributions.html

```python
from  scipy import stats, integrate
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
#Set a color map, cmap, object. 
cmap = sns.cubehelix_palette(n_colors=12, as_cmap=True, dark=0, light=1, reverse=True)
```

```python
#Create some data
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
```

```python
sns.jointplot(x="x", y="y", data=df);#kind can be hex, kde, 
```

```python
sns.kdeplot(df.x, df.y, cmap=cmap, n_levels=60, shade=True);
```

```python
plt.hist2d(df.x,df.y,bins=50)
```

```python
#So you can also create rvs to implement into these plots.
X = stats.norm.rvs(loc=0,scale=1,size=10)
Y = stats.norm.rvs(loc=0,scale=.5,size=10)
sns.kdeplot(X, Y, cmap=cmap, n_levels=60, shade=True);
```

```python
#This can also be implemented into a joint plot.
sns.jointplot(X,Y)
```

```python
# This is the 3D plotting toolkit
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Bivariate distribution using mlab...still no clear way to use this API
#to plot a customised rv distribution though. 
q = np.linspace(-5, 5, 200)
r = q
Q,R = np.meshgrid(q,r)
Z = plt.mlab.bivariate_normal(Q, R)

#Useful to examine the parameters for this method. 
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(Q,R, Z, cmap=cmap, rstride=3, cstride=3, linewidth=0)
```

```python
A = stats.norm.rvs(loc=0,scale=1,size=200)
B = stats.norm.rvs(loc=1,scale=0.5,size=200)
sns.jointplot(A,B)
```

## Sympy and Stats

Sympy has a very useful set of tools for examing symbolic elements of distributions. http://docs.sympy.org/dev/modules/stats.html There appear to be some holes, however, in terms of implementations (e.g., Poisson doesn't seem able to have the .dict method) and some of the methods (sample, where) are not clearly exampled or explained. 

```python
from sympy import symbols, integrate, expand_func,simplify,init_session, init_printing
from sympy.stats import Beta, density, E, P, variance
init_printing()
alpha, beta = symbols('alpha beta',positive=True)
z = symbols('z')
X = Beta('x',alpha,beta)
D = density(X)(z)
D
```

```python
expand_func(simplify(E(X,meijerg=True)))
```
