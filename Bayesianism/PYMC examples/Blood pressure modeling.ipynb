## PyMC analysis

So, it would be interesting to know if BP has changed as a function of time, noting that at a point in time there was a change in medication. The basic model is that BP is viewed as normally distributed. I'm not convinced this is the best distribution as the tails are problematic. For example, there will be a very very small chance that BP is less than 0, where in reality this is nonsense. Setting that aside for now, the next step is to consider the parameters. So there are two possible distributions (pre- and post-) mixed in the data observed. Each will have a mean and SD. There are therefore 4 parameters to consider. This begs the question of what parameters specify the mean and SD. In the case of the mean, the Gamma distribution is probably best, with hyper-parameters set to create a reasonable but imperfect distribution. What will be the most relevant statistic is the distribution of the difference between means. To enhance this, create a deterministic where instead we calculate an effect size, defined as difference between means divided by pooled variance. 

http://stats.stackexchange.com/questions/46626/fitting-model-for-two-normal-distributions-in-pymc

There are two apparent ways of measuring a combined systolic and diastolic cycle. 

First, mean arterial pressure (MAP), which is $MAP \simeq DP -\frac{1}{3}(SP - DP)$. 

The second is pulse pressure, defined as $PP \approx SP - DP$. 

In considering a model of blood pressure functioning, there following parameters apply, specifying from the model down. 

1. A combined metric varies between $0 ( DP = SP$ ) to $BP_{max}$.

2. The deterministic model is either $MAP$ or $PP$.

3. Variables DP and SP could be modeled as $DP/SP \sim Norm(\mu, \sigma)$. It is possibly better to consider that the range of BP is at 0 to a maximum (say 200??), and so a Gamma distribution may be preferable, with parameters shape $k$ and scale $\theta = \frac{1}{\beta}$.  The PDF is $\frac{1}{\Gamma(k) \theta^k} x^{k \,-\, 1} e^{-\frac{x}{\     theta}}$  https://en.wikipedia.org/wiki/Gamma_distribution

4. Hyper-parameters $\mu$ and $\sigma$ can take normalised distribution values, preferably with some limits. For a Gamma distribution, it seems it would be ideal for the expected value to equal something around a mean value for the $BP$ index. E[X] = kθ = α/β. 

5. There is probably a pre- and post- Indicator variable for the impact of an external, for example medication initiated, change.As it may not be known when the change occurs, one might need to model a value for $BP(t)$ to express $P(BP(t) = T)$. This might be represented as $\tau \sim Uniform(a, b)$, given we would have no knowledge or expectation when the change occurred in a randomised trial. 

```python
#Create two measures of combined BP
df['PP'] = df['Sys']-df['Dia']
df['MAP'] = df['Dia']- 1./3*(df.Sys - df.Dia)
#Convert these to np.arrays
PP = df.PP.values
MAP = df.MAP.values
#Convert sys and Dia to array
Systolic_obs = df.Sys.values
Diastolic_obs = df.Dia.values
```

As a reference https://github.com/CamDavidsonPilon/Probabilistic-Programming-and-Bayesian-Methods-for-Hackers/blob/master/Chapter2_MorePyMC/Chapter2.ipynb

```python
#Create hyperparameters.
mu = pm.Uniform('mu',100,180) #Not sure about this - assuming I don't know the mean Sys BP
mu2=pm.Uniform('mu2',70,100) 
#sigma = 1./(pm.Uniform('sigma',10,40))**2  #So assuming the BPis not varying by > +/- 40mmHg
sigma = pm.Uniform('sigma',.9,1.1) #This keeps E(mu) from 91-164mmHG, reasonable for this patient                       
@pm.deterministic
def tau(sigma=sigma):
    return 1./sigma**2
#Create variables for SP and DP. 
SP = pm.Gamma('SP',mu,tau) #Not sure that BP is normally distributed as never <0 or >200
DP = pm.Gamma('DP',mu2,tau)#Maybe gamma, but start with normal. 
#Write the deterministic model
@pm.deterministic
def MAP_(Systolic=SP, Diastolic=DP):
    return Diastolic-(Systolic-Diastolic)/3.
```

```python
#Examine the returned variables
print 'Parameter values are; mu {0}, mu2 {1} SP {2} and DP {3}'.format(mu.value,mu2.value,SP.value,DP.value)
print
print 'The value of the MAP deterministic: '
print MAP_.value
```

```python
#Create observations and a model
observation_SP = pm.Normal("observation_SP", SP, tau=sigma, value=Systolic_obs, observed=True)
observation_DP = pm.Normal("observation_DP", DP, tau=sigma, value=Diastolic_obs, observed=True)
model = pm.Model([observation_SP,observation_DP, MAP_,SP,DP,mu,mu2,sigma])
#Run the MCMC process
iterations = 100000
burnin = 10000
thin=1
mcmc = pm.MCMC(model)
mcmc.sample( iterations, burn=burnin, thin=thin)
```

```python
#Create the trace arrays for each parameter
SP_samples = mcmc.trace('SP')[:]
DP_samples = mcmc.trace('DP')[:]
MAP_samples = mcmc.trace('MAP_')
#Inspect the results
from pymc.Matplot import plot
plot(mcmc)
```

```python
Expected_Sys = SP_samples.mean()
Expected_Dia = DP_samples.mean()
print 'The expected Systolic is {0:0.0f} and Diastolic is {1:0.0f}'.format(Expected_Sys, Expected_Dia)
STD_Sys = SP_samples.std()
STD_Dia = DP_samples.std()
print 'The STD for Systolic is {0:0.2f} and Diastolic is {1:0.2f}'.format(STD_Sys, STD_Dia)
```
