# Heart rate changes as a function of medication

```python
from IPython.core.pylabtools import figsize
import numpy as np
from matplotlib import pyplot as plt
figsize(11, 9)
import scipy.stats as stats
import pymc as pm
```

http://nbviewer.ipython.org/github/CamDavidsonPilon/Probabilistic-Programming-and-Bayesian-Methods-for-Hackers/blob/master/Chapter1_Introduction/Chapter1.ipynb

```python
#Create data list of resting HR from Fitbit
count_data = [58,59,58,58,59,61,62,
              63,62,61,59,57,56,57,
              57,58,59,61,61,61,62,
              63,63,63,63,63,63,61,
              60,59,59,62,63,63,64,#2-8Aug
              64,64,65,65,66,66,64,
              66,66,65,66,65,64,67,
              64,65,64,63,64,61,60,#to 29 August
              59,57,57,59,60,58,58,
              57,57,59,58,59,59,58,
              59,59,59,61,61,61,63,
              63,62,64,63,63,63,63,
              62,61,60,58,55,55,54,
              55,54,54,55,57,56,55,
              55,56,58,59,59]
```

```python
#Inspect the data
figsize(12.5, 3.5)
n_count_data = len(count_data)
plt.bar(np.arange(n_count_data), count_data,
        color="#348ABD", alpha=0.4,width=.5)
plt.xlabel("Time (days)")
plt.ylabel("Resting HR")
plt.title("Resting HR variation")
plt.grid(1)
plt.xlim(0, n_count_data)

print 'Mean HR is {0:0.0f} with SD = {1:0.2f}'.format(np.mean(count_data),np.std(count_data))
```

```python
# Set up parameters
alpha = 1.0 / np.mean(count_data)  # Recall count_data is the
                               # variable that holds our txt counts
lambda_1 = pm.Exponential("lambda_1", alpha)
lambda_2 = pm.Exponential("lambda_2", alpha)
tau = pm.DiscreteUniform("tau", lower=0, upper=n_count_data)

@pm.deterministic
def lambda_(tau=tau, lambda_1=lambda_1, lambda_2=lambda_2):
    out = np.zeros(len(count_data))
    out[:tau] = lambda_1  # lambda before tau is lambda1
    out[tau:] = lambda_2  # lambda after (and including) tau is lambda2
    return out
#Examine the returned variables
print 'Parameter values are; Tau {0}, Lambda1 {1} and Lambda2 {2}'.format(tau.value, lambda_1.value, lambda_2.value)
print
print 'The value of the lambda array: '
print lambda_.value
```

```python
#Create observations and a model
observation = pm.Poisson("obs", lambda_, value=count_data, observed=True)
model = pm.Model([observation, lambda_1, lambda_2, tau])
#Run the MCMC process
iterations = 100000
burnin = 10000
thin=1
mcmc = pm.MCMC(model)
mcmc.sample( iterations, burn=burnin, thin=thin)
```

```python
#Create the trace arrays for each parameter
lambda_1_samples = mcmc.trace('lambda_1')[:]
lambda_2_samples = mcmc.trace('lambda_2')[:]
tau_samples = mcmc.trace('tau')[:]
#Inspect the results
from pymc.Matplot import plot
plot(mcmc)
```

```python
#Visualise the posterior distribution
figsize(12.5, 5)
# tau_samples, lambda_1_samples, lambda_2_samples contain
# N samples from the corresponding posterior distribution
N = tau_samples.shape[0]
expected_HR = np.zeros(n_count_data)
for day in range(0, n_count_data):
    # ix is a bool index of all tau samples corresponding to
    # the switchpoint occurring prior to value of 'day'
    ix = day < tau_samples
    # Each posterior sample corresponds to a value for tau.
    # for each day, that value of tau indicates whether we're "before"
    # (in the lambda1 "regime") or
    #  "after" (in the lambda2 "regime") the switchpoint.
    # by taking the posterior sample of lambda1/2 accordingly, we can average
    # over all samples to get an expected value for lambda on that day.
    # As explained, the "HR count" random variable is Poisson distributed,
    # and therefore lambda (the poisson parameter) is the expected value of
    # "HR count".
    expected_HR[day] = (lambda_1_samples[ix].sum()
                                   + lambda_2_samples[~ix].sum()) / N


plt.plot(range(n_count_data), expected_HR, lw=4, color="#E24A33",
         label="Resting HR")
plt.xlim(0, n_count_data)
plt.xlabel("Day")
plt.ylabel("Expected HR")
plt.title("HR as a function of medication change")
plt.vlines(53,35,80,linestyle="dashed",colors='purple',label="Sotolol")
plt.ylim(35, 80)
plt.bar(np.arange(len(count_data)), count_data, color="#348ABD", alpha=0.65,
        label="Observed HR")
plt.grid(True)
plt.legend(loc="upper left")
plt.savefig('RestingHR.png')
```

## Exercise HR

```python
#Create data list of exercise HR from Fitbit
count_data2 = [124,128,127,
               126,126,
               128,126,130,
               119,130,
               130,125,
               127,
               111,90,109,90,109,
               109,111,116,108]
```

```python
#Inspect the data
figsize(12.5, 3.5)
n_count_data2 = len(count_data2)
plt.bar(np.arange(n_count_data2), count_data2,
        color="#348ABD", alpha=0.4,width=.5)
plt.xlabel("Time (days)")
plt.ylabel("Exercise (Stairclimber) HR")
plt.title("HR variation")
plt.grid(1)
plt.xlim(0, n_count_data2)

print 'Mean HR is {0:0.0f} with SD = {1:0.2f}'.format(np.mean(count_data2),np.std(count_data2))
```

```python
# Set up parameters
alpha2 = 1.0 / np.mean(count_data2)  # Recall count_data is the
                               # variable that holds our txt counts
lambda_1_2 = pm.Exponential("lambda_1_2", alpha2)
lambda_2_2 = pm.Exponential("lambda_2_2", alpha2)
tau2 = pm.DiscreteUniform("tau2", lower=0, upper=n_count_data2)
@pm.deterministic
def lambda_(tau=tau2, lambda_1=lambda_1_2, lambda_2=lambda_2_2):
    out = np.zeros(len(count_data2))
    out[:tau] = lambda_1  # lambda before tau is lambda1
    out[tau:] = lambda_2  # lambda after (and including) tau is lambda2
    return out

#Examine the returned variables
print 'Parameter values are; Tau {0}, Lambda1 {1} and Lambda2 {2}'.format(tau2.value, lambda_1_2.value, lambda_2_2.value)
print
print 'The value of the lambda array: '
print lambda_.value
```

```python
#Create observations and a model
observation2 = pm.Poisson("obs", lambda_, value=count_data2, observed=True)
model = pm.Model([observation2, lambda_1_2, lambda_2_2, tau2])
#Run the MCMC process
iterations2 = 100000
burnin2 = 10000
thin2=1
mcmc2 = pm.MCMC(model)
mcmc2.sample( iterations2, burn=burnin2, thin=thin2)
```

```python
#Create the trace arrays for each parameter
lambda_1_samples2 = mcmc2.trace('lambda_1_2')[:]
lambda_2_samples2 = mcmc2.trace('lambda_2_2')[:]
tau_samples2 = mcmc2.trace('tau2')[:]
#Inspect the results
from pymc.Matplot import plot
plot(mcmc2)
```

```python
#Visualise the posterior distribution
figsize(12.5, 5)
# tau_samples, lambda_1_samples, lambda_2_samples contain
# N samples from the corresponding posterior distribution
N2 = tau_samples2.shape[0]
expected_HR2 = np.zeros(n_count_data2)
for day in range(0, n_count_data2):
    # ix is a bool index of all tau samples corresponding to
    # the switchpoint occurring prior to value of 'day'
    ix = day < tau_samples2
    # Each posterior sample corresponds to a value for tau.
    # for each day, that value of tau indicates whether we're "before"
    # (in the lambda1 "regime") or
    #  "after" (in the lambda2 "regime") the switchpoint.
    # by taking the posterior sample of lambda1/2 accordingly, we can average
    # over all samples to get an expected value for lambda on that day.
    # As explained, the "HR count" random variable is Poisson distributed,
    # and therefore lambda (the poisson parameter) is the expected value of
    # "HR count".
    expected_HR2[day] = (lambda_1_samples2[ix].sum()
                                   + lambda_2_samples2[~ix].sum()) / N2


plt.errorbar(range(n_count_data2), expected_HR2, yerr=expected_HR2.std(),
             capthick=2,lw=2, color="#E24A33",ecolor='g',fmt='o',label="Exercise HR")
plt.xlim(0, n_count_data2)
plt.xlabel("Day")
plt.ylabel("Expected HR")
plt.title("HR as a function of medication change")
plt.vlines(13,35,140,linestyle="dashed",lw=4,colors='purple',label="Sotolol")
plt.ylim(35, 140)
plt.bar(np.arange(len(count_data2)), count_data2, color="#348ABD", alpha=0.65,
        label="Observed HR",align='center')
plt.grid(True)
plt.legend(loc="center left")
plt.savefig('ExerciseHR.png',format='png')
```

```python
d2 = lambda_2_samples2.mean()-lambda_1_samples2.mean()/np.sqrt(lambda_2_samples2.std()+lambda_1_samples2.std())
print 'Mean effect size is {0:0.2f} for a mean difference in HR of {1:0.0f}'.format(d2,
                                                                                    lambda_2_samples2.mean()-lambda_1_samples2.mean())
```
