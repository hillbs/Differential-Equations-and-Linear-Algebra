## Example: Laplace's Rule of succession

Given $p$, which is unknown, there is a probability $Bern(p)$ that an event (sun rising) will occur again tomorrow. Bayesian paradigm suggests we treat $p$ as a r.v. (What does this mean?)

Laplace said, $p \sim Unif(0,1)$, the prior. $S_n = X_1 + X_2 + ... X_n$.

The structure is that $S_n|p \sim Bin(n,p); p \sim Unif(0,1)$.

The problem is to find the posterior distribution, $p|S_n.$ This is asking how many $S_n$'s are ones. 

And, what is $P(X_{n+1} = 1|S_n = n)?$ That is, the sun has risen for $n$-times in the last days, so what is the probability it will rise again tomorrow at time $n + 1?$

How do we find this posterior $distribution?$

$f(p|S_n = k)$. The sun has risen on $k$ of the last days. This is just Bayes' rule, written thus: 

$$f(p|S_n = k) = \frac{P(S_n = k|p) f(p)}{P(S_n = k)}$$

The denominator DOES NOT DEPEND on $p$. Defining it directly using the Law of Total Probability, one finds that: 
$$P(S_n = k) = \int_0^1 P(S_n = k|p) f(p) dp.$$

The denominator is a constant that doesn't depend on $p.$ So can be ignored. Also, the numerator does not depend on p, so the "n choose p" term is a constant, can be dropped in favour of proportionality. So the equation reduces to: 

$$f(p|S_n = k)\  \propto \ p^k (1 - p)^{n - k}$$

But, we know that the sun does rise, or has risen, every day. So the $q$-term is 0, leaving:

$$f(p|S_n = n) = (n + 1)p^n$$

This gives, for the probability of the posterior: 

$$P(X_{n+1} = 1|S_n = n) = \int_0^1 p(n+1)p^n dp$$

This is the expected value of this variable with this distribution. 

The answer is $\frac{n+1}{n+2}.$
