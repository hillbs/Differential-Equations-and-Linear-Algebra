# A Simulation of risk levels by group for Programme Delivery pre-treatment

```python
#p( offending ) = ROCROI
#p(non-offending ) = 1 - p(off)
#range = .4-.7
#so can be any 10 person random combination of the above.
#So this gives a single trial an array of random numbers between these values.
#
import random
import itertools
import numpy as np

def Generate_Group(n = 10):
    ''' Creates a random group of participants with .4 <RoC*RoI < .7'''
    Group_array = []
    mean_desistance = 0
    for num in range(n):
        Group_array.append(round(np.random.uniform(.4,.7),4))
        
    mean_desistance = np.random.uniform(0.4,0.7)
    return mean_desistance

def num_groups_pa(n = 150):        
    group = []
    for g in range(n):
        group.append(Generate_Group())
    return group    
    
def Analysis():
    result = num_groups_pa()
    ave = np.mean(result)
    stdevi = np.std(result)
    Hi = np.max(result)
    Lo = np.min(result)
    #print result
    print 'For a sample of 150 groups of 10 participants'
    print 'RoC*RoI between 0.4-0.7, randomly allocated'
    print 'The average risk over all those groups is', round(ave, 2)
    print 'The STD across the 150 groups is', round(stdevi,2)
    print 'The highest average risk per group is', round(Hi, 3),' & lowest is', round(Lo,3)
    return result
    
y = Analysis()
```

```python
%matplotlib inline
title("Distribution of risk scores\n N=150 groups, n=10 participants")
hist(y, bins=6, color='g')
xlabel("RoC*RoI bands")
ylabel("Frequency (0-150)")
text(.48,25,"Test")
```

So this assumes that the groupings are random. I am not sure this is an entirely correct assumption. Similarly, I am not sure that the distribution of risk levels by RoC*RoI is gaussian. This raises an issue of skewing outcomes by initial membership selection. One would expect this violation of normality of distribution, however, to weight toward a false-positive effect. 

IF one uses the Andrews & Bonta model of effect size (the BESD model, which OVERestimates more often than not), then intuitively, some conclusions arise from the above simulation:

1. One would expect to find an average risk level of .55 +/- .08 for 2/3 of the samples. It would be interesting to see if indeed the RQ is around this level for RoC*RoI.

2. An effect size from TREATMENT, however that is defined, would need to be greater than variability associated with random sampling of participants with varying risk levels.

3. Intuitively, one can see that comparing between any two groups has the potential to show a clearly spurious effect size. So it is that there is some reasonable probability that comparison, as in the RQ, could for lower sample sizes (of groups, in particular) be comparing random assignment rather than treatment effect.

Limitations of this are that:

1. RoC*RoI is not in fact a definitive measure of individual risk. So could also flatten risk to be equal for all members (removing a prior probability, but also the point of the risk estimation paragdigm).

2. The simulation necessarily assumes equality of impact on each group. So therapist impact, for example, is not available to the simulation. If the 0.3ish effect size of this was added to some, not all groups, a further consideration would therefore be comparing SD and covariance between RQ results.
