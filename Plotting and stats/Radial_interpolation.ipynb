# Interpolating with Radial Basis Functions

```python
x = linspace(-3,3,100)
```

```python
plot(x, exp(-1 * x **2))
t = title("Gaussian")
```

```python
plot(x, sqrt(1 + x **2))
t = title("Multiquadric")
```

```python
plot(x, 1. / sqrt(1 + x **2))
t = title("Inverse Multiquadric")
```

```python
plot(x, 1. / (1 + x **2))
t = title("Inverse Quadric")
```
